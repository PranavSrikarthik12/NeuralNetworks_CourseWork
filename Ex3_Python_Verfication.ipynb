```python
import numpy as np

# Sigmoid function
def sigmoid(x):
    return 1/(1+np.exp(-x))

# Given values
x = np.array([[1,0]])
y = 1

W1 = np.array([[0.2,0.3],
               [0.4,0.1]])

W2 = np.array([[0.5],
               [0.6]])

lr = 0.1

# ---------- Forward Pass ----------
hidden_input = np.dot(x, W1)
hidden_output = sigmoid(hidden_input)

output_input = np.dot(hidden_output, W2)
prediction = sigmoid(output_input)

loss = 0.5 * (y - prediction)**2

print("Hidden Layer Output:")
print(hidden_output)

print("\nPrediction:")
print(prediction)

print("\nLoss:")
print(loss)

# ---------- Backpropagation ----------
delta_output = (prediction - y) * prediction * (1 - prediction)

grad_W2 = hidden_output.T @ delta_output

delta_hidden = (delta_output @ W2.T) * hidden_output * (1 - hidden_output)

grad_W1 = x.T @ delta_hidden

# ---------- Weight Update ----------
W2_new = W2 - lr * grad_W2
W1_new = W1 - lr * grad_W1

print("\nUpdated W1")
print(W1_new)

print("\nUpdated W2")
print(W2_new)
```
